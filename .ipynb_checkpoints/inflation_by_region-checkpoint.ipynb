# Diving into the June Inflation Numbers
The headline CPI inflation rate in June was reported to be 1.9% at the national level. Out of curiosity, it might be interesting to take a look at the inflation rate by region as well as by type of consumer good. So, in this notebook, we'll visualize the aggregate and commodity-level CPI inflation by region and see what we find. To do this, we'll be using the handy [geopandas](https://github.com/geopandas/geopandas) package since it makes plotting things on a map much easier.

Table 8 is originally contained in a PDF file, so I had to do some preliminary processing (which I won't go into detail here; just use any PDF conversion tool you like) to turn it into a workable CSV file, here named 'june_inflation_by_region.csv'.

## Plotting Regional Inflation
Before we can plot the regional inflation rates on a Philippine map, we need to do a couple of things first in order to make sure that the regional inflation numbers do correspond to the correct region. One way to do this is to join (or merge) the `regions` geodataframe and the `inflation` on some common column (preferrably a column with unique rows).

### Joining (Merging) the Inflation Dataframe with the Regions Geodataframe
With pandas (and, by extension, geopandas), merging our two dataframes is as simple as calling the `merge()` function. The problem here, as we can see above, is that the two dataframes don't have a common column with unique rows. So let's create such a column.

Looking at the two dataframes, we can see that the `Area` column in `inflation` is probably a good candidate. Let's add a column called `Area` to the `regions` (geo)dataframe and populate this with rows whose values can be matched with those under the `inflation.Area` column. We do this by defining a helper function:

```python
def make_area_col(row):
    text = row['REGION']
    if 'Manila' in text:
        return u'NCR'
    else:
        return text[text.rfind('(')+1: text.rfind(')')]
```

All this function does is extract the string inside the parentheses of each row under `regions.REGION`. So, using this function, we get:

```python
regions['Area'] = regions.apply(lambda x: make_area_col(x), axis=1)
```

```python
regions['Area']
```

Now, we can merge `regions` and `inflation` as follows:

```python
reg_infl = regions.merge(inflation, on='Area')
```

```python
type(reg_infl)
```

We use the `regions` geodataframe to call the `merge()` function so that `regions` is the left dataframe. According to the [relevant geopandas docs](http://geopandas.org/mergingdata.html), this ensures that the `merge()` returns a geodataframe object (recall that `inflation` is a regular pandas dataframe). This is shown when we call `type()` on `reg_infl`. Also, the `merge()` function performs an inner join unless otherwise indicated in its `how` keyword argument.

Let's see if `reg_infl` is really what we expect it to be:

```python
reg_infl.shape
```

```python
reg_infl.columns
```

```python
reg_infl[['REGION', 'Area', 'All Items']]
```

This looks exactly as we intend it to be, although `reg_infl` doesn't include the row for the NIR entries (since, once again, this region has not yet been included in the shape file we're using here). With that aside, we're just about ready to plot this on a map.

### Plotting Regional Inflation Rates for All Items
Let's first test the waters a little bit by plotting the regional CPI inflation rates for all items in the Pinoy household's basket of goods.

```python
% matplotlib inline
plt.style.use('ggplot')

ax = reg_infl.plot(column='All Items', linewidth=0, cmap='Reds')

ax.set_title('Inflation Rates by Region, All Items')
ax.set_axis_off()

mx = reg_infl['All Items'].max()
mn = reg_infl['All Items'].min()
sm = plt.cm.ScalarMappable(
    cmap='Reds',
    norm=plt.Normalize(vmin=mn, vmax=mx)
)
sm._A = []
plt.colorbar(sm, ax=ax, shrink=0.5)

plt.savefig('Output\\01 - All Items.png', bbox_inches='tight')
```

Not bad at all. In the above plot, the darker shades of red indicate higher inflation rate, and you can see that there's quite a bit of variability in the YoY CPI change when we look at it on a regional basis. We see the highest rates of inflation last month in limited areas of Luzon, but are particularly pronounced in large portions of Visayas and Mindanao. It's worth pointing out that, apart from the small recovery in oil prices last month, the lingering effects of the recent El Nino on agriculture also contributed to the [uptick in June consumer prices](http://www.sunstar.com.ph/manila/local-news/2016/07/05/inflation-rises-19-june-483512), and no one else in the Philippines [felt the dry spell more sharply](http://www.philstar.com/headlines/2016/05/23/1586110/el-nino-have-lingering-effects) than farmers in Mindanao.

### Plotting Regional CPI Inflation by Commodity Group
To map the rest of the commodity-level inflation rates, we'll be modifying our code above as follows (we could've included 'All Items' here instead of separately above, but we wanted to see what the map would look like with a single example first, so don't hate me for it):

```python
from textwrap import wrap       # To take care of long titles

# to ease the monotony, we use different sequential cmaps
cmaps = ['Blues', 'BuGn', 'BuPu',
         'GnBu', 'Greens', 'Greys', 'Oranges', 'OrRd',
         'afmhot', 'autumn', 'bone', 'cool']

for i, col in enumerate(reg_infl.columns[4:]):
    ax = reg_infl.plot(column=col, linewidth=0, cmap=cmaps[i])
    
    title = 'Inflation Rates by Region, %s' % col
    ax.set_title('\n'.join(wrap(title, 40)))
    ax.set_axis_off()

    mx = reg_infl[col].max()
    mn = reg_infl[col].min()
    sm = plt.cm.ScalarMappable(
        cmap=cmaps[i],
        norm=plt.Normalize(vmin=mn, vmax=mx)
    )
    sm._A = []
    plt.colorbar(sm, ax=ax, shrink=0.5)

    plt.savefig('Output\\%02d - %s.png' % (i+2, col),
                bbox_inches='tight')
```
